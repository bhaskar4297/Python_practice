```python
from pyspark.sql.functions import lit            #df is mutable
people_df=spark.read.csv('/content/abcde.csv',inferSchema=True,header=True)
#people_df.show()
people_df=people_df.withColumn('new age' ,people_df["age"]-5).withColumn('team' ,lit('India'))
people_df=people_df.withColumnRenamed('new age', 'debut age')
people_df.filter(lit(1)==lit(2)).show()
people_df.show()
```

```python
people_df.createOrReplaceTempView('people')
spark.sql("select * from people where `debut age` < 30").show()
```
